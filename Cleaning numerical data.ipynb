## Data cleaning

```python
data.columns = [data.columns[i].title().replace(' ', '_') for i in range(len(data.columns))]
data
```

```python
data['Customer_Lifetime_Value'] = data['Customer_Lifetime_Value'].astype('float')
```

```python
data.dtypes
```

```python
data['Effective_To_Date'] = pd.to_datetime(data['Effective_To_Date'])
#data.dtypes
data
```

```python
#Check NaN values per column
for i in data.select_dtypes('object', 'int'):
    print(i,data[i].isna().sum(), '\n')
```
